# **1. Business Problem Understanding**

## **1.1. Context** 

The types of financial products used by the public are more varied. One financial product that is well-known to the public is term deposits. The term deposit mechanism is that a customer deposits a sum of money at a bank or a financial institution, and the money can only be withdrawn after a certain period of time. As compensation, customers will be given a fixed interest according to the nominal amount of money deposited.

Nevertheless, as business entities with financial products and respective customers, banks still have to compete to not lose customers. One of the ways to acquire new customers is by conducting a marketing campaign.

However, marketing campaign must be designed as efficient as possible in order to gain profit, which could help the bank to survive, compete with the competitors, and even expand their business. 

Banks need deposits to fund their lending activities, manage liquidity, and maintain operational efficiency. Deposits provide a low-cost source of capital that banks use to create credit, earn interest income, and meet regulatory requirements. They also help establish trust and stability, attract business, and generate additional revenue through fees. Overall, deposits are essential for a bank's financial health and its ability to support economic growth. Term deposit subscriptions directly impact the bank's revenue and financial stability, as they are a source of funds that the bank can use for lending and investments. 

Predicting term deposit subscriptions allows the bank to focus its marketing efforts on the most promising customer, thus improving the return on investment (ROI) of marketing campaigns. By identifying customers who are likely to subscribe, the bank can tailor its marketing messages and offers, enhancing customer satisfaction and loyalty. Customers who subscribe to term deposits are often more valuable in the long run, as they tend to have a higher lifetime value through continued engagement with the bank's products and services.

Machine Learning can be important for identifying the customer who are likely to subscribe, by analyzing the data of the customer, to understand the customer behavior, and because of that, the machine learning models able to predict customer that have potential to make a term deposit. It is important to accurately allocate the budget for market campaign, by knowing the potential customer. This capability allows banks to focus their marketing efforts on individuals with a higher tendency to subscribe, thereby increasing the effectiveness of their campaigns.

## **1.5. Evaluation Metric**

The evaluation metric that will be used for the Model will be:
1. Accuracy
Accuracy measures the proportion of total correct predictions (both subscribers and non-subscribers) out of all predictions made. . It gives a general idea of how well the model is performing overall in predicting whether a customer will subscribe to a term deposit.
2. Precision
"It measures the proportion of actual subscribers among all customers predicted to subscribe. It is useful for minimizing the targeting of customers who are not actually interested. Precision helps to ensure when the model predicts a customer will subscribe, they are likely to do so."

3. Recall
Recall measures the proportion of actual subscribers that the model successfully identifies. It is crucial to ensure that most of the customers who are actually interested in subscribing are captured by the model.
4. F1-Score
The F1-score is a balanced measure that combines precision and recall. Useful to balance between capturing as many actual subscribers as possible (recall) and ensuring that the customers predicted to subscribe are indeed likely to do so (precision).
5. Confusion Matrix
The confusion matrix provides a detailed view of prediction performance, showing how many subscribers were correctly identified (True Positives), how many non-subscribers were incorrectly predicted as subscribers (False Positives), and how many subscribers were missed (False Negatives). It helps in understanding the modelâ€™s errors in detail.

## **3.3. Outliers**

Normality test performed so that the outliers can be handled appropriately.

```python
import scipy.stats as stats
# List of numerical columns to analyze
numerical_columns = ['age', 'balance', 'campaign', 'pdays']

# Function to plot histogram and Q-Q plot 
def plot_normality(bank_df, column):
    # Histogram
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(bank_df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
    # Q-Q Plot
    plt.subplot(1, 2, 2)
    stats.probplot(bank_df[column], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {column}')
    
    plt.tight_layout()
    plt.show()

# Apply the function to each numerical column
for col in numerical_columns:
    plot_normality(bank_df, col)

# Statistical Normality Tests
for col in numerical_columns:
    # Shapiro-Wilk Test
    stat, p_value = stats.shapiro(bank_df[col].dropna())
    print(f'Shapiro-Wilk Test p-value for {col}: {p_value}')
    
    # Kolmogorov-Smirnov Test
    stat, p_value = stats.kstest(bank_df[col].dropna(), 'norm', args=(bank_df[col].mean(), bank_df[col].std()))
    print(f'Kolmogorov-Smirnov Test p-value for {col}: {p_value}')
```

From the graphic and the results, indicates that the data is not normally distributed. The numerical column such as the age, balance, number of contacts performed during this campaign and for this client, number of days after the client was contacted from the previous campaign show it is right-skewed distribution. This distribution could affect the the machine learning in terms of learning patterns


```python
# Function to calculate outliers using IQR
def calculate_outliers(bank_df):
    outlier_counts = {}
    for column in bank_df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = bank_df[column].quantile(0.25)
        Q3 = bank_df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = bank_df[(bank_df[column] < lower_bound) | (bank_df[column] > upper_bound)]
        outlier_counts[column] = len(outliers)
    return outlier_counts

# Calculate outliers
outlier_counts = calculate_outliers(bank_df)

# Convert to Series for better readability
outlier_counts_series = pd.Series(outlier_counts, name='Outlier Counts')
print(outlier_counts_series)
```

```python
# To inspect 
def inspect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Inspect outliers for each numerical column
numerical_columns = bank_df.select_dtypes(include=['int64', 'float64']).columns
outliers_dict = {col: inspect_outliers(bank_df, col) for col in numerical_columns}

# Print outlier values
for col, outliers in outliers_dict.items():
    print(f"\nOutliers in column '{col}':")
    print(outliers[col])
```

```python
# For column with categorical data
def analyze_categorical_column(bank_df, column_name, threshold_percentage=1.0):
    """
    Analyzes the frequency of categories in a categorical column and identifies rare categories.
    
    :param df: DataFrame containing the data
    :param column_name: Column name to analyze
    :param threshold_percentage: Threshold percentage to identify rare categories
    :return: Frequency counts, percentages, and rare categories
    """
    # Count frequency of each category
    freq_counts = bank_df[column_name].value_counts()
    
    # Calculate percentage of each category
    percentage = (freq_counts / len(bank_df)) * 100
    
    # Identify rare categories based on the threshold percentage
    rare_categories = percentage[percentage < threshold_percentage].index.tolist()
    
    return freq_counts, percentage, rare_categories

# List of categorical columns to analyze
categorical_columns = ['job', 'housing', 'loan', 'contact', 'month', 'poutcome']

# Analyze each categorical column
for col in categorical_columns:
    freq_counts, percentage, rare_categories = analyze_categorical_column(bank_df, col)
    print(f"Column: {col}")
    print(f"Frequency Counts:\n{freq_counts}\n")
    print(f"Percentage:\n{percentage}\n")
    print(f"Rare Categories (less than threshold): {rare_categories}\n")
```

By training the model on one subset (training set) and testing it on another (testing set), we can get an estimate of the model's performance on new data.

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(random_state=42))
])

# Fit the pipeline on the training data
pipeline.fit(X_train, y_train)

# Predict on the test data
y_pred = pipeline.predict(X_test)
```

```python
# Base models
logreg = LogisticRegression(solver='liblinear', random_state=42)
knn = KNeighborsClassifier()
dtree = DecisionTreeClassifier(random_state=42)

# Ensemble models - Stacking
sc = StackingClassifier(estimators=[
    ('dtree', dtree),
    ('knn', knn),
    ('lr', logreg)
])

# Ensemble models - Soft Voting
vc = VotingClassifier(estimators=[
    ('dtree', dtree),
    ('knn', knn),
    ('lr', logreg)
], voting='soft')

# Ensemble models - Boosting
ada = AdaBoostClassifier(random_state=42)
gbc = GradientBoostingClassifier(random_state=42)
xgb = XGBClassifier(random_state=42, verbosity=0)

# Ensemble models - Bagging
rf = RandomForestClassifier(random_state=42)

# List of models 
models = [
    ('Logistic Regression', logreg),
    ('KNN', knn),
    ('Decision Tree', dtree),
    ('Stacking Classifier', sc),
    ('Voting Classifier', vc),
    ('AdaBoost', ada),
    ('Gradient Boosting', gbc),
    ('XGBoost', xgb),
    ('Random Forest', rf)
]
```

```python
f2_scorer = make_scorer(fbeta_score, beta=2)
results = {}

for name, model in models:
    # Create a pipeline for each model
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    # Perform cross-validation
    model_cv = cross_val_score(pipeline, X_train, y_train, cv=5, scoring=f2_scorer, error_score='raise')
    
    # Store the mean score
    results[name] = model_cv.mean()

# Output the results
for name, score in results.items():
    print(f"{name}: {score:.4f}")
```

```python
# Predict probabilities
y_prob = pipeline.predict_proba(X_test)[:, 1]

# Predict binary outcomes based on a threshold
y_pred = (y_prob > 0.5).astype(int)  
y_test_series = pd.Series(y_test)
y_pred_series = pd.Series(y_pred)

# Calculate conversion rate
conversion_rate = y_test_series.value_counts(normalize=True).get(1, 0)  # Actual positive rate
predicted_conversion_rate = y_pred_series.value_counts(normalize=True).get(1, 0)  # Predicted positive rate

print(f"Actual Conversion Rate: {conversion_rate:.2f}")
print(f"Predicted Conversion Rate: {predicted_conversion_rate:.2f}")
```

```python
from sklearn.metrics import recall_score
# Calculate true positive rate (recall for 'yes')
recall_yes = recall_score(y_test, y_pred, pos_label=1)

# Calculate false positive rate
false_positive_rate = 1 - recall_score(y_test, y_pred, pos_label=0)

# Calculate costs
cost_per_false_positive = 100  # Example value
cost_per_false_negative = 50  # Example value
total_cost_false_positives = false_positive_rate * cost_per_false_positive * len(y_test)
total_cost_false_negatives = (1 - recall_yes) * cost_per_false_negative * len(y_test)

# Calculate expected revenue
revenue_per_correct_prediction = 200  # Example value
total_revenue = (recall_yes * len(y_test) * revenue_per_correct_prediction)

# Calculate ROI
roi = (total_revenue - (total_cost_false_positives + total_cost_false_negatives)) / (total_cost_false_positives + total_cost_false_negatives) * 100

print("Conversion Rate:", conversion_rate)
print("Predicted Conversion Rate:", predicted_conversion_rate)
print("True Positive Rate (Recall for 'yes'):", recall_yes)
print("False Positive Rate:", false_positive_rate)
print("Total Cost of False Positives:", total_cost_false_positives)
print("Total Cost of False Negatives:", total_cost_false_negatives)
print("Expected ROI:", roi ,"%")
```
